## 5. Modeling

### 5.1 Random Forest

```python

num_cols = X.select_dtypes(include=['int64', 'float64']).columns
cat_cols = X.select_dtypes(include=['object', 'category']).columns

encoder = OneHotEncoder(drop='first', handle_unknown='ignore')
scaler = StandardScaler()

preprocessor = ColumnTransformer(
    transformers=[
        ('num', scaler, num_cols),
        ('cat', encoder, cat_cols)
    ]
)

# Build and train pipeline
pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42))
])

pipeline.fit(X_train, y_train)

# Evaluate
y_pred = pipeline.predict(X_test)
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

<b>Interpretation:</b>

- Accuracy = 0.88 â€“ The model performs well overall, correctly classifying most employees.
- Precision (1) = 0.92 â€“ Very high; when the model predicts an employee will leave, it is almost always correct.
- Recall (1) = 0.31 â€“ Low; the model still misses a notable number of employees who actually leave.
- F1 (1) = 0.46 â€“ Moderate; it shows a good balance between precision and recall, but thereâ€™s room for improvement in detecting all leavers.

<b>Conclusion:</b> The model is reliable in predicting leavers when it does so, but it could be improved to capture more actual cases of attrition (higher recall).

```python
# Checking class imbalance
y.value_counts(normalize=True)
```

```python
# Lets use the class weights to tell the model to pay more attention to the minority class

pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('classifier', RandomForestClassifier(random_state=42, class_weight='balanced'))
])

pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix
print(classification_report(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
```

<b>Interpretation:</b>
After adding class weights, the model:
- Accuracy = 0.85 â€“ Slightly lower overall accuracy, but this is expected after adjusting for class imbalance.
- Precision (1) = 0.62 â€“ Decent; when the model predicts an employee will leave, it is often correct.
- Recall (1) = 0.13 â€“ Very low; the model still misses most employees who actually leave.
- F1 (1) = 0.21 â€“ Low; it shows the model struggles to consistently identify leavers.

<b>Conclusion:</b> Adding class weights helped the model pay slightly more attention to the minority class (employees who leave), but itâ€™s still not effective at detecting them. Further tuning (e.g., different class weights, oversampling, or other algorithms) may be needed to improve recall without sacrificing too much precision.
